```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("Everything installed successfully!")
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 data points
X = np.random.rand(100, 1) * 10  # Values between 0 and 10
noise = np.random.randn(100, 1)  # Random noise

# Define linear relationship: y = 2x + 3 + noise
y = 2 * X + 3 + noise

# Quick check
print(X[:5], y[:5])
```

```python
# Plotting the data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label='Data points')
plt.title('Synthetic Linear Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Create Linear Regression model instance
model = LinearRegression()

# Train the model with the synthetic data
model.fit(X, y)
```

```python
# Predict y using your trained model
y_pred = model.predict(X)

# Evaluate performance
mse = np.mean((y - y_pred) ** 2)  # Mean Squared Error
rmse = np.sqrt(mse)               # Root Mean Squared Error
r2 = model.score(X, y)            # R-squared score

print(f"Slope (Coefficient): {model.coef_[0][0]:.2f}")
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"RMSE: {rmse:.4f}")


# Plot the original data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label='Actual Data')

# Plot the predicted line
plt.plot(X, y_pred, color='red', linewidth=2, label='Model Prediction')

plt.title('Linear Regression Model Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import joblib

# Export your trained Linear Regression model
joblib.dump(model, 'linear_regression_model.pkl')
```
